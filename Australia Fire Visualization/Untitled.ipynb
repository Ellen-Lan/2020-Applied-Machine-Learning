```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(2, 2)
ax11, ax21, ax12, ax22 = axes.ravel()
ax11.set_title("ax11")
ax21.set_title("ax21")
ax12.set_title("ax12")
ax22.set_title("ax22")
```

```python
sin=np.sin(np.linspace(-4,4,100))
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(sin)
axes[0, 1].plot(sin, c='r')
```
