```python
from sklearn import datasets
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Activation
from sklearn.model_selection import train_test_split
from keras import regularizers
from keras.wrappers.scikit_learn import KerasClassifier, KerasRegressor
from sklearn.model_selection import GridSearchCV
import pandas as pd 
import numpy as np
```

#Task 1 [10 Points]
Run a multilayer perceptron (a densely connected feed forward neural network) with two
hidden layers and rectified linear nonlinearities on the digits dataset from sklearn using the
keras Sequential interface . Include code for selecting L2 regularization strength and number of
hidden units using GridSearchCV and evaluation on an independent test-set.

```python
#Load the digits dataset
digits = datasets.load_digits()
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, random_state = 0, test_size = 0.2, stratify = digits.target) 
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
def simple_model(kernel_reg=0.001, hidden_units=32, optimizer = 'adam'):
  model = Sequential([
      Dense(32, input_shape=(64,), activation='relu', kernel_regularizer = regularizers.l2(kernel_reg)), #input shape must match X_train
      Dense(10, activation='softmax')])
  model.compile(optimizer=optimizer, loss="categorical_crossentropy", metrics=['accuracy'])
  return model
```

```python
#running a gridsearch 
param_grid = {
    'epochs': [1, 5, 10], 
    'kernel_reg': [0.001, 0.01, 0.1, 1, 10], 
    'hidden_units': [32, 64, 256]}
clf = KerasClassifier(simple_model)
grid = GridSearchCV(clf, param_grid=param_grid, return_train_score = True)
grid.fit(X_train, y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
res = pd.DataFrame(grid.cv_results_)
res.pivot_table(index=["param_epochs", "param_kernel_reg", "param_hidden_units"],
                values=["mean_train_score", "mean_test_score"])  
```

```python
#evaluation on test set 
grid.score(X_test, y_test)
```
